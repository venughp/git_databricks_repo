```python
from pyspark.sql.functions import broadcast
spark.conf.set("spark.sql.adaptive.enabled", "false")
```

```python
df_fact=spark.createDataFrame(
    [
        (1, "US", 1000),
        (2, "CA", 2000),
        (3, "US", 3000),
        (4, "CA", 4000)
    ],["id", "country", "sales"]
)

df_dim = spark.createDataFrame(
    [
        ("US", "United States"),
        ("CA", "Canada")
    ],["country", "country_name"]
)
display(df_fact)
display(df_dim)
```

```python
df_join = df_fact.join(df_dim, df_fact.country == df_dim.country, "inner")
```

```python
df_join.display()
```

```python
df_join_broadcast = df_fact.join(broadcast(df_dim), df_fact.country == df_dim.country, "inner")
```

```python
df_join_broadcast.display()
```
