```python
from pyspark.sql.functions import broadcast
spark.conf.set("spark.sql.adaptive.enabled", "false")
```

```python
df_fact=spark.createDataFrame(
    [
        (1, "US", 1000),
        (2, "CA", 2000),
        (3, "US", 3000),
        (4, "CA", 4000)
    ],["id", "country", "sales"]
)

df_dim = spark.createDataFrame(
    [
        ("US", "United States"),
        ("CA", "Canada")
    ],["country", "country_name"]
)
display(df_fact)
display(df_dim)
```

```python
# temp view Vs Global View 
df_fact.createOrReplaceTempView("fact_trans")
df_dim.createOrReplaceTempView("dim_country")
```

```python
df_sql=spark.sql('''Select * 
from fact_trans ft
join dim_country dc
on ft.country = dc.country''')
```

```python
df_sql.display()
```

```python
df_sql_broadcast=spark.sql('''Select /*+broadcast(dc)*/ 
*                           
from fact_trans ft
join dim_country dc
on ft.country = dc.country''')
```

```python
df_sql_broadcast.display()
```

```python
df_sql_broadcast.explain(True)
```
